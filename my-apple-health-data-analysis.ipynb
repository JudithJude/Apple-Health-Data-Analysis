```python
#converting timezone to WAT
convert_tz = lambda x: x.to_pydatetime().replace(tzinfo=pytz.utc).astimezone(pytz.timezone("Africa/Lagos"))
the_year = lambda x: convert_tz(x).year
the_month = lambda x: '{}-{:02}'.format(convert_tz(x).year, convert_tz(x).month) #inefficient
the_date = lambda x: '{}-{:02}-{:02}'.format(convert_tz(x).year, convert_tz(x).month, convert_tz(x).day) #inefficient
the_day = lambda x: convert_tz(x).day
the_hour = lambda x: convert_tz(x).hour
the_minute = lambda x: convert_tz(x).minute
the_day_of_week = lambda x: convert_tz(x).weekday()
```

```python
# parsing date and time for each dataset (startDate)
steps["startDate"] = pd.to_datetime(steps["startDate"])
steps["year"] = steps["startDate"].map(the_year)
steps["month"] = steps["startDate"].map(the_month)
steps["date"] = steps["startDate"].map(the_date)
steps["day"] = steps["startDate"].map(the_day)
steps["hour"] = steps["startDate"].map(the_hour)
steps["dow"] = steps["startDate"].map(the_day_of_week)

sleep["startDate"] = pd.to_datetime(sleep["startDate"])
sleep["year"] = sleep["startDate"].map(the_year)
sleep["month"] = sleep["startDate"].map(the_month)
sleep["date"] = sleep["startDate"].map(the_date)
sleep["day"] = sleep["startDate"].map(the_day)
sleep["hour"] = sleep["startDate"].map(the_hour)
sleep["dow"] = sleep["startDate"].map(the_day_of_week)

walking_speed["startDate"] = pd.to_datetime(walking_speed["startDate"])
walking_speed["year"] = walking_speed["startDate"].map(the_year)
walking_speed["month"] = walking_speed["startDate"].map(the_month)
walking_speed["date"] = walking_speed["startDate"].map(the_date)
walking_speed["day"] = walking_speed["startDate"].map(the_day)
walking_speed["hour"] = walking_speed["startDate"].map(the_hour)
walking_speed["dow"] = walking_speed["startDate"].map(the_day_of_week)

energy_burnt["startDate"] = pd.to_datetime(energy_burnt["startDate"])
energy_burnt["year"] = energy_burnt["startDate"].map(the_year)
energy_burnt["month"] = energy_burnt["startDate"].map(the_month)
energy_burnt["date"] = energy_burnt["startDate"].map(the_date)
energy_burnt["day"] = energy_burnt["startDate"].map(the_day)
energy_burnt["hour"] = energy_burnt["startDate"].map(the_hour)
energy_burnt["dow"] = energy_burnt["startDate"].map(the_day_of_week)

dist_walk_run["startDate"] = pd.to_datetime(dist_walk_run["startDate"])
dist_walk_run["year"] = dist_walk_run["startDate"].map(the_year)
dist_walk_run["month"] = dist_walk_run["startDate"].map(the_month)
dist_walk_run["date"] = dist_walk_run["startDate"].map(the_date)
dist_walk_run["day"] = dist_walk_run["startDate"].map(the_day)
dist_walk_run["hour"] = dist_walk_run["startDate"].map(the_hour)
dist_walk_run["dow"] = dist_walk_run["startDate"].map(the_day_of_week)
```

```python
# parsing date and time for each dataset (endDate)
steps["endDate"] = pd.to_datetime(steps["endDate"])
steps["year2"] = steps["endDate"].map(the_year)
steps["month2"] = steps["endDate"].map(the_month)
steps["date2"] = steps["endDate"].map(the_date)
steps["day2"] = steps["endDate"].map(the_day)
steps["hour2"] = steps["endDate"].map(the_hour)
steps["dow2"] = steps["endDate"].map(the_day_of_week)

sleep["endDate"] = pd.to_datetime(sleep["endDate"])
sleep["year2"] = sleep["endDate"].map(the_year)
sleep["month2"] = sleep["endDate"].map(the_month)
sleep["date2"] = sleep["endDate"].map(the_date)
sleep["day2"] = sleep["endDate"].map(the_day)
sleep["hour2"] = sleep["endDate"].map(the_hour)
sleep["dow2"] = sleep["endDate"].map(the_day_of_week)

walking_speed["endDate"] = pd.to_datetime(walking_speed["endDate"])
walking_speed["year2"] = walking_speed["endDate"].map(the_year)
walking_speed["month2"] = walking_speed["endDate"].map(the_month)
walking_speed["date2"] = walking_speed["endDate"].map(the_date)
walking_speed["day2"] = walking_speed["endDate"].map(the_day)
walking_speed["hour2"] = walking_speed["endDate"].map(the_hour)
walking_speed["dow2"] = walking_speed["endDate"].map(the_day_of_week)

energy_burnt["endDate"] = pd.to_datetime(energy_burnt["endDate"])
energy_burnt["year2"] = energy_burnt["endDate"].map(the_year)
energy_burnt["month2"] = energy_burnt["endDate"].map(the_month)
energy_burnt["date2"] = energy_burnt["endDate"].map(the_date)
energy_burnt["day2"] = energy_burnt["endDate"].map(the_day)
energy_burnt["hour2"] = energy_burnt["endDate"].map(the_hour)
energy_burnt["dow2"] = energy_burnt["endDate"].map(the_day_of_week)

dist_walk_run["endDate"] = pd.to_datetime(dist_walk_run["endDate"])
dist_walk_run["year2"] = dist_walk_run["endDate"].map(the_year)
dist_walk_run["month2"] = dist_walk_run["endDate"].map(the_month)
dist_walk_run["date2"] = dist_walk_run["endDate"].map(the_date)
dist_walk_run["day2"] = dist_walk_run["endDate"].map(the_day)
dist_walk_run["hour2"] = dist_walk_run["endDate"].map(the_hour)
dist_walk_run["dow2"] = dist_walk_run["endDate"].map(the_day_of_week)
```

```python
#getting my sleep time
sleep["sleep_time"] = sleep["endDate"] - sleep["startDate"]
sleep.head()
```

```python
#grouping each dataset by date
steps_by_date = steps.groupby("date")["value"].sum().reset_index(name = "steps")
sleep_by_date = sleep.groupby("date")["sleep_time"].sum().reset_index(name = "sleep_time")
walking_speed = walking_speed.groupby("date")["value"].sum().reset_index(name = "walk_km/hr")
energy_burnt = energy_burnt.groupby("date")["value"].sum().reset_index(name = "kcal")
dist_walk_run = dist_walk_run.groupby("date")["value"].sum().reset_index(name = "walk_run_km")
```

```python
sleep_by_date.head()
```

```python
all_data_2 = pd.concat([steps_by_date, sleep_by_date, walking_speed, energy_burnt, dist_walk_run], ignore_index = True)
```

```python
steps_by_date.to_csv("steps_by_date.csv", index = False)
sleep_by_date.to_csv("sleep_by_date.csv", index = False)
walking_speed.to_csv("walkspeed_by_date.csv", index = False)
energy_burnt.to_csv("energy_burnt_by_date.csv", index = False)
dist_walk_run.to_csv("walkrun_by_date.csv", index = False)
```
